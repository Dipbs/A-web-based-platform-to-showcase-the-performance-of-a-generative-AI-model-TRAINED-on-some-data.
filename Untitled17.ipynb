```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments

# Install wandb if not already installed
!pip install wandb

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
# Add padding token to tokenizer
tokenizer.pad_token = tokenizer.eos_token
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Load and preprocess your dataset
# Assuming your dataset is in a file called 'my_dataset.txt'
from datasets import load_dataset
datasets = load_dataset('text', data_files={'train': 'Health index2.csv'})

# Tokenize the dataset and add labels
def tokenize_function(examples):
    # Tokenize the text
    tokenized = tokenizer(examples["text"], padding="max_length", truncation=True)
    # Add labels - shifted input_ids for language modeling
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_datasets = datasets.map(tokenize_function, batched=True)

# ... rest of your training code

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    save_steps=10_000,
    logging_dir="./logs",
    # The following line disables wandb logging if you don't want to use it
    report_to = "none" # This line will completely disable wandb
)

# The following lines are not needed if you disabled wandb using report_to = "none"
# import os
# os.environ["WANDB_DISABLED"] = "true"

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
)

```

```python
import gradio as gr

def generate_text(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=50)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

iface = gr.Interface(fn=generate_text,
                     inputs="text",
                     outputs="text",
                     title="My Trained GPT-2",
                     description="A fine-tuned generative model demo")

iface.launch()
```
